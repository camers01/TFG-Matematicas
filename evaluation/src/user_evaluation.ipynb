## User Evaluation

In this notebook, we will start by randomly choosing 5 cases from the seed case base in order to run a user evaluation. For each of those cases, we will print the image (graph to explain) and two texts (the Qwen (if XAI graph) or Idefics (if Statistical graph) zero-shot baseline and the Pixtral 'rag_solution' insights). We selected Qwen for XAI and Idefics for Statistical as the baselines for this comparison because they where the best performers for each graph category. 

We will attach each image and the corresponding two texts onto a Google Form, aiming to see if human preference aligns with the quantitative metrics obtained during offline evaluation by asking users to choose the the most useful option based on their criteria.

```python
import os
import pandas as pd
from IPython.display import display, Markdown, Image

# Path Setup
SRC_DIR = os.getcwd()
EVAL_ROOT = os.path.dirname(SRC_DIR)
PROJECT_ROOT = os.path.dirname(EVAL_ROOT)
XAI_RAG_DIR = os.path.join(PROJECT_ROOT, "graph_cbr_rag")
CASE_BASE_CSV = os.path.join(XAI_RAG_DIR, "data", "processed", "case_base.csv")
IMAGES_DIR = os.path.join(XAI_RAG_DIR, "data", "raw", "images")
RAG_OUTPUT_CSV = os.path.join(EVAL_ROOT, "data", "raw", "pixtral_rag_insights.csv")

# Load and merge Data
df_base = pd.read_csv(CASE_BASE_CSV, dtype={'id': str})
df_base['id'] = df_base['id'].str.zfill(6)
df_rag = pd.read_csv(RAG_OUTPUT_CSV, dtype={'id': str})
df_rag['id'] = df_rag['id'].str.zfill(6)
# Using inner join we filter to only keep the 530 successful cases
df_master = pd.merge(df_base, df_rag, on="id", how="inner")

# Select 5 Random Cases (we will use a fixed seed for replicabilty of choice)
RANDOM_SEED = 22
df_sample = df_master.sample(n=5, random_state=RANDOM_SEED)

for index, row in df_sample.iterrows():

    case_id = row['id']
    category = str(row['graph_category']).upper() # To choose which baseline to use

    # Get the image path
    img_path = os.path.join(IMAGES_DIR, f"{case_id}.png")

    # Dynamic zero-shot baseline selection
    if category == 'XAI':
        baseline_name = "QWEN"
        baseline_text = row['qwen_insight']
    else:
        baseline_name = "IDEFICS"
        baseline_text = row['idefics_insight']
    
    # Extract the few-shot RAG text from Pixtral
    rag_text = row['insight']
    
    print(f"========================================================================================================")
    print(f"CASE ID: {case_id} | CATEGORY: {category} | COMPARED AGAINST: {baseline_name}")

    if os.path.exists(img_path):
        display(Image(filename=img_path, width=800))
    else:
        print(f"[Warning: Image not found at {img_path}]")
    
    markdown_content = f"""
**>>> OPTION A: ZERO-SHOT {baseline_name} BASELINE**
> {baseline_text}

**>>> OPTION B: PIXTRAL FEW-SHOT RAG (MATH TFG)**
> {rag_text}
<br>
"""
    display(Markdown(markdown_content))
```

As a second part of this evaluation, we will compare the evolution of our current CBR-RAG architecture against the previous iteration developed in the Computer Science project. To ensure a perfectly symmetric and fair comparison, we isolated the 50 Explainable AI (XAI) cases used for this Mathematics project, matching them across both databases using a unique fingerprint combination with columns present in both datasets (`domain`, `graph_type`, and `variables`).

We randomly sample 5 cases, for which human evaluators will be presented with the XAI graph alongside two unlabelled explanations: Option A and Option B (representing the previous system and the newly implemented pipeline, randomized to prevent position bias). Through this blind test, users will be asked to select the explanation they find most useful.

```python
# Path setup for the old CS project files
OLD_CASES_CSV = os.path.join(EVAL_ROOT, "data", "raw", "original_50_xai_cases.csv")
OLD_RAG_CSV = os.path.join(EVAL_ROOT, "data", "raw", "pixtral_rag_insights_xai.csv")

# 1. RECONSTRUCT THE OLD CS PROJECT CASES
# Load metadata and RAG insights, forcing string types for IDs
df_old_meta = pd.read_csv(OLD_CASES_CSV, dtype={'id': str})
df_old_rag = pd.read_csv(OLD_RAG_CSV, dtype={'id': str})
df_old_meta['id'] = df_old_meta['id'].str.zfill(6)
df_old_rag['id'] = df_old_rag['id'].str.zfill(6)
# Merge both old dataframes on their shared 'id' to get the text aligned with the metadata
df_old_complete = pd.merge(df_old_meta, df_old_rag, on='id', how='inner')

# 2. MATCHING CS DATAFRAME WITH MATH DATAFRAME
fingerprint_cols = ['domain', 'graph_type', 'variables']
# Keep only the fingerprint and the old generated text ('insight' from pixtral_rag_insights_xai.csv)
df_old_subset = df_old_complete[fingerprint_cols + ['insight']].rename(
    columns={'insight': 'old_cbr_insight'}
)
# Filter the master dataframe from Cell 1 to only look at XAI cases
df_master_xai = df_master[df_master['graph_category'].str.upper() == 'XAI']
# Final Merge: Match the reconstructed old cases with the new case base using the strict fingerprint
df_xai_comparison = pd.merge(df_master_xai, df_old_subset, on=fingerprint_cols, how="inner")
print(f"Match successful! Found {len(df_xai_comparison)} complete XAI cases from the old project in the new Case Base.\n")

# 3. SELECT 5 RANDOM CASES
RANDOM_SEED_XAI = 72
df_sample_xai = df_xai_comparison.sample(n=5, random_state=RANDOM_SEED_XAI)

# 4. VIEW THE TEXTS COMPARISON
print("--- XAI COMPARISON: OLD CBR (CS) vs NEW CBR (MATH) ---\n")
for index, row in df_sample_xai.iterrows():
    new_case_id = row['id']
    img_path = os.path.join(IMAGES_DIR, f"{new_case_id}.png")
    # Extract the texts
    old_cbr_text = row['old_cbr_insight'] # From Computer Science TFG
    new_cbr_text = row['insight']         # From Mathematics TFG
    print(f"========================================================================================================")
    print(f"NEW CASE ID: {new_case_id} | GRAPH TYPE: {row['graph_type']} | DOMAIN: {row['domain']}")
    if os.path.exists(img_path):
        display(Image(filename=img_path, width=800))
    else:
        print(f"[Warning: Image not found at {img_path}]")
    markdown_content = f"""
**>>> OPTION A: PREVIOUS CBR SYSTEM (Computer Science TFG)**
> {old_cbr_text}

**>>> OPTION B: NEW CBR SYSTEM (Mathematics TFG)**
> {new_cbr_text}
<br>
"""
    display(Markdown(markdown_content))
```
